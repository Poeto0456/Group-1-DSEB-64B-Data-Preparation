```python
def process_credit_data(df):
    # Tạo bản sao để không thay đổi dữ liệu gốc
    processed_df = df.copy()
    
    # Xác định các dòng active và có debt, limit đều bằng 0
    zero_mask = ((processed_df['CREDIT_ACTIVE'].isin(['Active', 'Bad debt'])) & 
                 (processed_df['AMT_CREDIT_SUM_DEBT'] == 0) & 
                 (processed_df['AMT_CREDIT_SUM_LIMIT'] == 0))
    
    # Tìm các dòng có giá trị dương để tính tỉ lệ nợ
    positive_rows = processed_df[
        (processed_df['AMT_CREDIT_SUM'] > 0) & 
        (processed_df['AMT_CREDIT_SUM_DEBT'] > 0) & 
        (processed_df['AMT_CREDIT_SUM_LIMIT'] > 0)
    ]
    
    # Tính tỉ lệ nợ trung bình từ các dòng có giá trị dương
    avg_debt_ratio = (positive_rows['AMT_CREDIT_SUM_DEBT'] / positive_rows['AMT_CREDIT_SUM']).mean()
    
    # Áp dụng tỉ lệ nợ trung bình cho các dòng active có debt và limit bằng 0
    processed_df.loc[zero_mask, 'AMT_CREDIT_SUM_DEBT'] = \
        processed_df.loc[zero_mask, 'AMT_CREDIT_SUM'] * avg_debt_ratio
    
    # Làm tròn debt thành số nguyên
    processed_df.loc[zero_mask, 'AMT_CREDIT_SUM_DEBT'] = \
        processed_df.loc[zero_mask, 'AMT_CREDIT_SUM_DEBT'].round().astype(int)
    
    # Tính limit = sum - debt
    processed_df.loc[zero_mask, 'AMT_CREDIT_SUM_LIMIT'] = \
        processed_df.loc[zero_mask, 'AMT_CREDIT_SUM'] - \
        processed_df.loc[zero_mask, 'AMT_CREDIT_SUM_DEBT']
    
    return processed_df

bureau = process_credit_data(bureau)
```

```python
# Kiểm tra và tính lại các giá trị khi AMT_CREDIT_SUM == 0
def adjust_credit_sum(row):
    if row['AMT_CREDIT_SUM'] == 0:
        # Tính lại AMT_CREDIT_SUM từ AMT_CREDIT_SUM_DEBT và AMT_CREDIT_SUM_LIMIT
        row['AMT_CREDIT_SUM'] = row['AMT_CREDIT_SUM_DEBT'] + row['AMT_CREDIT_SUM_LIMIT']
    return row

# Áp dụng hàm điều chỉnh cho toàn bộ DataFrame
bureau = bureau.apply(adjust_credit_sum, axis=1)
```

```python
def adjust_loans(row):
    # Nếu AMT_CREDIT_SUM_DEBT < 0, gán về 0
    if (row['AMT_CREDIT_SUM_DEBT'] < 0) & (row['AMT_CREDIT_SUM_LIMIT'] != 0):
        row['AMT_CREDIT_SUM_DEBT'] = 0
    
    # Điều chỉnh AMT_CREDIT_SUM hoặc AMT_CREDIT_SUM_LIMIT
    if row['AMT_CREDIT_SUM'] == 0 and row['AMT_CREDIT_SUM_LIMIT'] != 0:
        row['AMT_CREDIT_SUM'] = row['AMT_CREDIT_SUM_LIMIT']
    elif row['AMT_CREDIT_SUM'] != 0 and row['AMT_CREDIT_SUM_LIMIT'] == 0:
        row['AMT_CREDIT_SUM_LIMIT'] = row['AMT_CREDIT_SUM']
    
    # Đảm bảo logic SUM = DEBT + LIMIT
    row['AMT_CREDIT_SUM'] = row['AMT_CREDIT_SUM_DEBT'] + row['AMT_CREDIT_SUM_LIMIT']
    
    return row

# Áp dụng hàm cho toàn bộ DataFrame
bureau = bureau.apply(adjust_loans, axis=1)
```
