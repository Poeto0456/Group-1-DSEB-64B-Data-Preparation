Tạo hàm tính tổng số tiền đã trả trong mỗi kỳ thanh toán

```python

def calculate_total_payment(group):
    total = 0
    total_list = []
    for _, row in group.iterrows():
        # Số tiền thanh toán trong khoảng hiện tại
        payment = row['AMT_PAYMENT']
        
        # Nếu tổng vượt AMT_INSTALMENT, chỉ lấy AMT_INSTALMENT
        if total + payment > row['AMT_INSTALMENT']:
            total = row['AMT_INSTALMENT']
        else:
            total += payment
        
        total_list.append(total)
    group['TOTAL_AMT_PAYMENT'] = total_list
    return group




```

```python
df_filtered = df_filtered.groupby(['SK_ID_PREV', 'NUM_INSTALMENT_NUMBER']).apply(calculate_total_payment)

# Kiểm tra kết quả
df_filtered
```

Xử lý trường hợp số tiền phải trả = 0 (AMT_INSTALL = 0)


```python
df_filtered['SK_ID_PREV'][df_filtered['AMT_INSTALMENT'] == 0].unique()
```

```python
# Bước 1: Lấy danh sách SK_ID_PREV từ điều kiện AMT_INSTALMENT == 0
selected_ids = df_filtered['SK_ID_PREV'][df_filtered['AMT_INSTALMENT'] == 0].unique()

# Bước 2: Xử lý từng trường hợp
for sk_id in selected_ids:
    # Lọc các hàng theo SK_ID_PREV
    sub_df = df_filtered[df_filtered['SK_ID_PREV'] == sk_id]
    
    # Trường hợp 1: Nếu có NUM_INSTALMENT_NUMBER = 1 và chứa NaN
    if (sub_df['NUM_INSTALMENT_NUMBER'] == 1).any() and sub_df[['AMT_PAYMENT', 'DAYS_ENTRY_PAYMENT']].isna().any().any():
        # Xóa các hàng chứa NaN
        df_filtered = df_filtered.drop(
            sub_df[(sub_df['NUM_INSTALMENT_NUMBER'] == 1) & (sub_df[['AMT_PAYMENT', 'DAYS_ENTRY_PAYMENT']].isna().any(axis=1))].index
        )
        
    else:
        # Trường hợp 2: Fill giá trị NaN bằng giá trị ngay trước nó
        sub_df = sub_df.fillna(method='ffill')

        # Cập nhật lại vào bảng gốc
        df_filtered.update(sub_df)
```

Xử lý trường hợp trả đủ nhưng do chia thành nhiều lần trả nên có giá trị nan

```python
# Bước 1: Tìm các hàng có giá trị NaN trong AMT_PAYMENT
nan_condition = df_filtered['AMT_PAYMENT'].isna()

# Bước 2: Dịch dữ liệu xuống 1 hàng để kiểm tra điều kiện của hàng ngay trước
shifted_amt_install = df_filtered['AMT_INSTALMENT'].shift(1)
shifted_amt_payment = df_filtered['AMT_PAYMENT'].shift(1)
shifted_total_amt_payment = df_filtered['TOTAL_AMT_PAYMENT'].shift(1)
shifted_sk_id_prev = df_filtered['SK_ID_PREV'].shift(1)

# Bước 3: Kiểm tra điều kiện cho hàng ngay trước
valid_prev_condition = (
    (shifted_amt_install < shifted_amt_payment) & 
    (shifted_amt_install == shifted_total_amt_payment) &
    (shifted_sk_id_prev == df_filtered['SK_ID_PREV'])  # Đảm bảo cùng SK_ID_PREV
)

# Bước 4: Kết hợp điều kiện để chọn SK_ID_PREV
df_filtered.loc[nan_condition & valid_prev_condition, 'SK_ID_PREV'].unique()
```

```python
selected_ids = df_filtered.loc[nan_condition & valid_prev_condition, 'SK_ID_PREV'].unique()

# Bước 5: Lọc và điền giá trị NaN trong các cột AMT_PAYMENT và DAYS_ENTRY_PAYMENT
df_filtered.loc[
    (df_filtered['SK_ID_PREV'].isin(selected_ids)) & (df_filtered['AMT_PAYMENT'].isna()), 
    'AMT_PAYMENT'
] = df_filtered['AMT_INSTALMENT']

df_filtered.loc[
    (df_filtered['SK_ID_PREV'].isin(selected_ids)) & (df_filtered['DAYS_ENTRY_PAYMENT'].isna()), 
    'DAYS_ENTRY_PAYMENT'
] = df_filtered['DAYS_INSTALMENT']
```
