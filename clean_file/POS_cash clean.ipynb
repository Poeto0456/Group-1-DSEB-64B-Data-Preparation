```python
filtered_ids = pos[
    (pos['CNT_INSTALMENT'].isna()) &
    (pos['CNT_INSTALMENT_FUTURE'].isna()) &
    (pos['NAME_CONTRACT_STATUS'].isin(['Returned to the store', 'Canceled']))
][['SK_ID_PREV', 'SK_ID_CURR']].drop_duplicates()

# Tạo mask cho tất cả các dòng có cặp ID cần xử lý
id_mask = pos.set_index(['SK_ID_PREV', 'SK_ID_CURR']).index.isin(
    filtered_ids.set_index(['SK_ID_PREV', 'SK_ID_CURR']).index
)

# Tạo bản sao của dataframe gốc chỉ với các dòng cần xử lý
returned_canceled = pos[id_mask].copy()

print("Số dòng trong bảng mới:", len(returned_canceled))
```

```python
# 5. Thay đổi cách xử lý để không làm mất index
def process_group(df):
    # Tạo temporary DataFrame để sắp xếp và xử lý
    temp_df = df.copy().sort_values(['SK_ID_PREV', 'SK_ID_CURR', 'MONTHS_BALANCE'])
    result_df = df.copy()
    
    for sk_id_prev in temp_df['SK_ID_PREV'].unique():
        group_mask = temp_df['SK_ID_PREV'] == sk_id_prev
        group = temp_df[group_mask]
        original_indices = group.index
        
        # Nếu dòng cuối nhóm là 'Returned to the store' hoặc 'Canceled'
        if group.iloc[-1]['NAME_CONTRACT_STATUS'] in ['Returned to the store', 'Canceled']:
            # Kiểm tra nếu dòng cuối cùng không có dòng sau
            if len(group) == 1 or group.iloc[-1]['MONTHS_BALANCE'] < group.iloc[-2]['MONTHS_BALANCE']:
                continue
            else:
                # Áp dụng bfill và cập nhật kết quả theo index gốc
                filled_group = group.bfill()
                result_df.loc[original_indices] = filled_group
        else:
            # Áp dụng bfill cho toàn nhóm
            filled_group = group.bfill()
            result_df.loc[original_indices] = filled_group
    
    return result_df

# 6. Áp dụng xử lý
returned_canceled = process_group(returned_canceled)
returned_canceled[returned_canceled.isnull().any(axis=1)]
```

```python
filtered_ids = pos[
    (pos['CNT_INSTALMENT'].isna()) &
    (pos['CNT_INSTALMENT_FUTURE'].isna()) &
    (pos['NAME_CONTRACT_STATUS'] == 'Active')
][['SK_ID_PREV', 'SK_ID_CURR']].drop_duplicates()

# Tạo mask cho tất cả các dòng có cặp ID cần xử lý
id_mask = pos.set_index(['SK_ID_PREV', 'SK_ID_CURR']).index.isin(
    filtered_ids.set_index(['SK_ID_PREV', 'SK_ID_CURR']).index
)

active = pos[id_mask].copy()
active= active.sort_values(by=['SK_ID_PREV', 'SK_ID_CURR', 'MONTHS_BALANCE'], ascending=True)

print("Số dòng trong bảng mới:", len(active))
```

```python
# Duyệt qua từng nhóm SK_ID_PREV
for sk_id_prev, group in active.groupby('SK_ID_PREV'):
    group_indices = group.index
    
    # Lấy giá trị ở dòng cuối của CNT_INSTALMENT
    last_cnt_instalment = group['CNT_INSTALMENT'].iloc[-1]
    
    # Thay thế toàn bộ CNT_INSTALMENT bằng giá trị cuối
    active.loc[group_indices, 'CNT_INSTALMENT'] = last_cnt_instalment
    
    # Tính toán CNT_INSTALMENT_FUTURE
    future_values = []
    current_value = last_cnt_instalment  # Bắt đầu từ giá trị cuối của CNT_INSTALMENT
    for i in range(len(group_indices) - 1, -1, -1):  # Duyệt ngược từ dòng cuối lên dòng đầu
        if i == len(group_indices) - 1:
            # Giá trị cuối cùng của CNT_INSTALMENT_FUTURE = CNT_INSTALMENT cuối
            future_values.append(current_value)
        else:
            # Trừ đi khoảng cách MONTHS_BALANCE
            balance_diff = group.loc[group_indices[i + 1], 'MONTHS_BALANCE'] - group.loc[group_indices[i], 'MONTHS_BALANCE']
            current_value -= balance_diff
            future_values.append(current_value)
    
    # Gán giá trị mới cho CNT_INSTALMENT_FUTURE
    active.loc[group_indices, 'CNT_INSTALMENT_FUTURE'] = future_values

active[active.isnull().any(axis=1)]
```

## Xử lý lỗi không đồng nhất CNT_INSTALMENT trong cùng 1 nhóm

```python
# Tạo một hàm tính mode cho nhóm
def calculate_mode(group):
    mode_value = group.mode()
    return mode_value.iloc[0] if not mode_value.empty else group.iloc[0]  # Giả sử nếu không có mode, chọn giá trị đầu tiên

# Nhóm theo SK_ID_PREV và SK_ID_CURR, sau đó áp dụng hàm calculate_mode cho CNT_INSTALMENT
pos['CNT_INSTALMENT'] = pos.groupby(['SK_ID_PREV', 'SK_ID_CURR'])['CNT_INSTALMENT'].transform(calculate_mode)

# Kiểm tra lại kết quả
print(pos.groupby(['SK_ID_PREV', 'SK_ID_CURR'])['CNT_INSTALMENT'].nunique())  # Tất cả các nhóm phải có 1 unique value
```
