## Cleaning AMT_ANNUITY column

```python
app[app['AMT_ANNUITY'].isna()]
```

### In all rows that this column is NaN, the result shows that the contract type is Cash loans. Therefore, we fillna in this column with the AMT_CREDIT divided by the average months of all the customers requiring Cash loans

```python
# Filter "Cash loans" customers
cash_loans = app[app['NAME_CONTRACT_TYPE'] == 'Cash loans']

# Remove NaN in AMT_CREDIT and AMT_ANNUITY
cash_loans_filtered = cash_loans.dropna(subset=['AMT_CREDIT', 'AMT_ANNUITY'])

# Calculate the average months
cash_loans_filtered['MONTHS'] = cash_loans_filtered['AMT_CREDIT'] / cash_loans_filtered['AMT_ANNUITY']
average_months = cash_loans_filtered['MONTHS'].mean()

print(f"S thng trung bnh cho hp ng 'Cash loans': {average_months:.2f}")
```

```python
# Fill the NaN value in AMT_ANNUITY
avg_months = average_months 
app['AMT_ANNUITY'] = app.apply(
    lambda row: row['AMT_CREDIT'] / avg_months if pd.isnull(row['AMT_ANNUITY']) and row['NAME_CONTRACT_TYPE'] == 'Cash loans' else row['AMT_ANNUITY'], 
    axis=1
)
```

## Cleaning AMT_GOODS_PRICE column

```python
app[app['AMT_GOODS_PRICE'].isna()]
```

### In all rows that this column is NaN, the result shows that the contract type is Revolving loans. Therefore, we fillna in this column with the median value where all the customers requiring Revolving loans

```python
# Median of AMT_GOODS_PRICE when NAME_CONTRACT_TYPE = "Revolving loans"
revolving_median = app.loc[app['NAME_CONTRACT_TYPE'] == 'Revolving loans', 'AMT_GOODS_PRICE'].median()

# Fillna by the median value calculated above
app.loc[(app['NAME_CONTRACT_TYPE'] == 'Revolving loans') & (app['AMT_GOODS_PRICE'].isna()), 'AMT_GOODS_PRICE'] = revolving_median
```

## Cleaning NAME_TYPE_SUITE column

```python
# Unique family status filtered that NAME_TYPE_SUITE is NaN
app[app['NAME_TYPE_SUITE'].isna()]['NAME_FAMILY_STATUS'].value_counts()
```

```python
# Find the mode of NAME_TYPE_SUITE for each group of NAME_FAMILY_STATUS
mode_by_family_status = app.groupby('NAME_FAMILY_STATUS')['NAME_TYPE_SUITE'].agg(lambda x: x.mode()[0] if not x.mode().empty else 'Unknown')
mode_by_family_status
```

```python
# Fill with the mode filtered for each NAME_FAMILY_STATUS group, as they have a domain relationship
app['NAME_TYPE_SUITE'] = app.apply(
    lambda row: mode_by_family_status[row['NAME_FAMILY_STATUS']] if pd.isna(row['NAME_TYPE_SUITE']) else row['NAME_TYPE_SUITE'],
    axis=1
)
```

## Cleaning OWN_CAR_AGE column

```python
# NaN rows in OWN_CAR_AGE
missing_car_age = app[app['OWN_CAR_AGE'].isna()]

# Check if these rows are the customers that don't own car (FLAG_OWN_CAR == 'N') or not
missing_car_age_no_car = missing_car_age[missing_car_age['FLAG_OWN_CAR'] == 'N']

# Check the result
print(f"S lng dng thiu OWN_CAR_AGE: {missing_car_age.shape[0]}")
print(f"S lng dng thiu OWN_CAR_AGE v khng s hu xe: {missing_car_age_no_car.shape[0]}")

missing_car_age_no_car
```

```python
app[(app['FLAG_OWN_CAR'] == 'Y') & (app['OWN_CAR_AGE'] == 0)]
```

### There are 162,188 NaN rows, of which 162,184 customers dont own a car. So for those who own a car, we fillna with median, and for those who do not, we fillna with -1

```python
# Fill -1 for those who don't own car
app.loc[app['FLAG_OWN_CAR'] == 'N', 'OWN_CAR_AGE'] = -1

# Fill median for those who own car
median_car_age = app.loc[app['FLAG_OWN_CAR'] == 'Y', 'OWN_CAR_AGE'].median()
app.loc[(app['FLAG_OWN_CAR'] == 'Y') & (app['OWN_CAR_AGE'].isna()), 'OWN_CAR_AGE'] = median_car_age
```

## Cleaning OCCUPATION_TYPE column

```python
# Divide AMT_INCOME_TOTAL into groups (quantile)
app['INCOME_QUANTILE'] = pd.qcut(app['AMT_INCOME_TOTAL'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])
```

```python
# Find the most popular value in each group
occupation_mode = app.groupby(['NAME_EDUCATION_TYPE', 'INCOME_QUANTILE'])['OCCUPATION_TYPE'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)
print(occupation_mode)
```

### Fill with the mode within groups based on NAME_EDUCATION_TYPE and INCOME_QUANTILE (low/medium/high/very high), as they have a domain relationship

```python
def fill_occupation(row):
    if pd.isna(row['OCCUPATION_TYPE']):
        return occupation_mode.get((row['NAME_EDUCATION_TYPE'], row['INCOME_QUANTILE']), row['OCCUPATION_TYPE'])
    else:
        return row['OCCUPATION_TYPE']

app['OCCUPATION_TYPE'] = app.apply(fill_occupation, axis=1)
```

## Cleaning CNT_FAM_MEMBERS column

```python
app[app['CNT_FAM_MEMBERS'].isna()]
```

### Only 1 NaN value in a customer whose family status is unknown, therefore fillna with 0

```python
app['CNT_FAM_MEMBERS'].fillna(value= 0, inplace= True)
```

## Cleaning EXT_SOURCE columns

```python
app[['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']].describe()
```

```python
# Fillna with the median as this comes from external information
for col in ['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']:
    app[col].fillna(app[col].median(), inplace=True)
```

## Cleaning the remaining columns

```python
# Find the other missing columns
missing_cols = app.columns[app.isna().any()].tolist()

# For other columns, fillna with the median where columns are numerical, otherwise fillna with the mode
for col in missing_cols:
    if app[col].dtype == 'object': 
        app[col] = app[col].fillna(app[col].mode()[0])
    else:  
        app[col] = app[col].fillna(app[col].median())  
```
