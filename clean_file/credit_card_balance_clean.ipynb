```python
# If all other methods are 0: distribute the total equally across the columns
condition1 = (credit['AMT_DRAWINGS_CURRENT'] > 0) & \
             (credit['AMT_DRAWINGS_ATM_CURRENT'] == 0) & \
             (credit['AMT_DRAWINGS_POS_CURRENT'] == 0) & \
             (credit['AMT_DRAWINGS_OTHER_CURRENT'] == 0)

credit.loc[condition1, ['AMT_DRAWINGS_ATM_CURRENT', 'AMT_DRAWINGS_POS_CURRENT', 'AMT_DRAWINGS_OTHER_CURRENT']] = \
    np.tile((credit.loc[condition1, 'AMT_DRAWINGS_CURRENT'] / 3).to_numpy(), (3, 1)).T
```

```python
# Check the result
credit[(credit['AMT_DRAWINGS_CURRENT'] > 0) & (credit['AMT_DRAWINGS_ATM_CURRENT'] == 0) & (credit['AMT_DRAWINGS_POS_CURRENT'] == 0) & (credit['AMT_DRAWINGS_OTHER_CURRENT'] > 0) &(abs(credit['AMT_DRAWINGS_CURRENT'] - 
            (credit['AMT_DRAWINGS_ATM_CURRENT'] + 
             credit['AMT_DRAWINGS_POS_CURRENT'] + 
             credit['AMT_DRAWINGS_OTHER_CURRENT'])) > 1e-6)]
```

```python
# If some methods have positive amounts but the sum is not equal: adjust the residual values proportionally

for col in ['AMT_DRAWINGS_ATM_CURRENT', 'AMT_DRAWINGS_POS_CURRENT', 'AMT_DRAWINGS_OTHER_CURRENT']:
    # Condition for each case
    condition2 = (credit['AMT_DRAWINGS_CURRENT'] > 0) & \
                 (credit[col] > 0) & \
                 (credit[[c for c in ['AMT_DRAWINGS_ATM_CURRENT', 'AMT_DRAWINGS_POS_CURRENT', 'AMT_DRAWINGS_OTHER_CURRENT'] if c != col]] == 0).all(axis=1) & \
                 (abs(credit['AMT_DRAWINGS_CURRENT'] - 
                      (credit['AMT_DRAWINGS_ATM_CURRENT'] + 
                       credit['AMT_DRAWINGS_POS_CURRENT'] + 
                       credit['AMT_DRAWINGS_OTHER_CURRENT'])) > 1e-6)

    # Calculate the remaining value
    remaining_value = credit.loc[condition2, 'AMT_DRAWINGS_CURRENT'] - credit.loc[condition2, col]

    # Assign value to the columns where amount is 0
    credit.loc[condition2, [c for c in ['AMT_DRAWINGS_ATM_CURRENT', 'AMT_DRAWINGS_POS_CURRENT', 'AMT_DRAWINGS_OTHER_CURRENT'] if c != col]] = \
        np.tile((remaining_value / 2).to_numpy(), (2, 1)).T
```

```python
# Adjust the residual values proportionally
for col1, col2, col3 in [
    ('AMT_DRAWINGS_ATM_CURRENT', 'AMT_DRAWINGS_POS_CURRENT', 'AMT_DRAWINGS_OTHER_CURRENT'),
    ('AMT_DRAWINGS_ATM_CURRENT', 'AMT_DRAWINGS_OTHER_CURRENT', 'AMT_DRAWINGS_POS_CURRENT'),
    ('AMT_DRAWINGS_POS_CURRENT', 'AMT_DRAWINGS_OTHER_CURRENT', 'AMT_DRAWINGS_ATM_CURRENT')
]:
    # Condition for each case
    condition3 = (credit['AMT_DRAWINGS_CURRENT'] > 0) & \
                 (credit[col1] > 0) & \
                 (credit[col2] > 0) & \
                 (credit[col3] == 0) & \
                 (abs(credit['AMT_DRAWINGS_CURRENT'] - 
                      (credit['AMT_DRAWINGS_ATM_CURRENT'] + 
                       credit['AMT_DRAWINGS_POS_CURRENT'] + 
                       credit['AMT_DRAWINGS_OTHER_CURRENT'])) > 1e-6)

    # Perform calculations
    credit.loc[condition3, col3] = (
        credit.loc[condition3, 'AMT_DRAWINGS_CURRENT'] - 
        credit.loc[condition3, col1] - 
        credit.loc[condition3, col2]
    )
```

## Check PAYMENT columns

```python
'''
AMT_INST_MIN_REGULARITY is the minimum installment due for the current month, while AMT_PAYMENT_CURRENT and AMT_PAYMENT_TOTAL_CURRENT 
represent the total amount to pay and the total paid in the previous month, respectively. 
Thus, AMT_INST_MIN_REGULARITY is calculated as the difference between the two payment values
'''

# Adjust installment value to be the difference between the two mentioned payment values for the prior month
credit = credit.sort_values(by=['SK_ID_PREV', 'MONTHS_BALANCE'], ascending=[True, False])

credit['AMT_PAYMENT_CURRENT_PREV'] = credit.groupby('SK_ID_PREV')['AMT_PAYMENT_CURRENT'].shift(-1)
credit['AMT_PAYMENT_TOTAL_CURRENT_PREV'] = credit.groupby('SK_ID_PREV')['AMT_PAYMENT_TOTAL_CURRENT'].shift(-1)

credit['AMT_INST_MIN_REGULARITY'] = credit['AMT_PAYMENT_CURRENT_PREV'] - credit['AMT_PAYMENT_TOTAL_CURRENT_PREV']
credit = credit.drop(columns=['AMT_PAYMENT_CURRENT_PREV', 'AMT_PAYMENT_TOTAL_CURRENT_PREV'])
```

```python
credit.fillna(value= 0, inplace= True)
```

### There are still some case when the minimum installment > 0 but payment values are 0, but the installment next month is 0, meaning that they had paid fully in the previous month but was recorded as 0. Thus, we set all the payment values in this case to be equal to the minimum installment

```python
credit.loc[
    (credit['AMT_PAYMENT_CURRENT'] == 0) & 
    (credit['AMT_PAYMENT_TOTAL_CURRENT'] == 0) & 
    (credit['AMT_INST_MIN_REGULARITY'] != 0), 
    ['AMT_PAYMENT_CURRENT', 'AMT_PAYMENT_TOTAL_CURRENT']
] = credit['AMT_INST_MIN_REGULARITY']
```

### As we can see, some rows in the CNT_INSTALLMENT_MATURE_CUM column doesn't follow the rule that if the payment was made, the count would increase but if not, the count would stay as the previous month. So we adjust this problem

```python
def adjust_cnt_installment(df):
    first_values = df.groupby('SK_ID_PREV')['CNT_INSTALMENT_MATURE_CUM'].transform('first')
    non_zero_payment = df['AMT_PAYMENT_TOTAL_CURRENT'] != 0
    cumulative_addition = non_zero_payment.groupby(df['SK_ID_PREV']).cumsum()
    df['CNT_INSTALMENT_MATURE_CUM'] = first_values + (cumulative_addition - cumulative_addition.groupby(df['SK_ID_PREV']).transform('first'))
    
    return df

credit = adjust_cnt_installment(credit)
```
